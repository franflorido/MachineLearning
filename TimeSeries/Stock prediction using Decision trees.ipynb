# Stock prediction using Time series analysis

In this programm we are going to use a decision tree algorithm to predict if the price of the stock will decrease o increase

## Importing libraries and data

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import numpy as np
import pandas as pd
```

```python
amazon = pd.read_csv('D:/Documentos/Master/temariomaster/p6.DescriptiveMachineLearning/Stockprediction/amzn.csv')
amazon
```

## Preprocess the data

```python
#set date as index
amazon = amazon.set_index(pd.DatetimeIndex(amazon['Date'].values))
amazon
```

```python
amazon.index.name = 'Date'
amazon
```

### Now we are going to create a column that will say if the price has increase or decrease, that is the label column

We are going to do that by creating a column that compares the prices of the past day to todays one, if the price is lower set 0 if it is higher set 1

```python
amazon['Price_Up'] = np.where(amazon['Close'].shift(-1) > amazon['Close'], 1, 0)
```

```python
amazon
```

```python
#check null values
amazon.isnull().values.any()
```

```python
#delete date column
del amazon['Date']
amazon
```

## Train and test the model

```python
# getting X and y data
X = amazon.iloc[:,0:6].values
X
```

```python
y = amazon.iloc[:,6].values
y
```

```python
# Getting train and testing data with 80-20% proportion
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2) 
```

```python
tree = DecisionTreeClassifier()
```

```python
tree.fit(X_train,y_train)
```

```python
print(tree.score(X_train,y_train))
```

```python
tree_predictions = tree.predict(X_test)
```

```python
print(tree_predictions)
```

```python
y_test
```

## Test and Train using grid search cross validation

```python
from sklearn.model_selection import StratifiedShuffleSplit
sss = StratifiedShuffleSplit(n_splits=1,test_size=0.3,random_state=13)
```

```python
for train_index, test_index in sss.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
```

```python
params = {'criterion':["gini", "entropy"], 
                                  'max_depth':[None,3,5,7,9],
                                  'min_samples_split':[2,4,8,16,20]}
```

```python
from sklearn.model_selection import GridSearchCV
miModelo = DecisionTreeClassifier()
gs = GridSearchCV(estimator=miModelo,param_grid=params,scoring='accuracy',cv=5,verbose=1)
```

```python
gs.fit(X_train,y_train)
```

```python
resultsCV = gs.cv_results_
resultsCV
```

```python
clfBest = gs.best_estimator_
clfBest
```

```python
clfBest.fit(X_train,y_train)
```

```python
y_pred = clfBest.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test,y_pred))
```
